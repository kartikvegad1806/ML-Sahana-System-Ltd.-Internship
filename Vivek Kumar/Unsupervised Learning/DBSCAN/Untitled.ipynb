```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
df = pd.read_csv("Clustering_gmm (2).csv")
df
x = df[['Weight','Height']]
plt.show()
db = DBSCAN(eps=0.8, min_samples=2)
df['Cluster'] = db.fit_predict(x)

plt.scatter(df['Weight'], df['Height'], c=df['Cluster'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("After DBSCAN Clustering")
plt.show()
```
